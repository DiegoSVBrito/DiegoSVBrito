```python

#alterando o padrão de ',' para '.' - Efeito melhora a visualização e posterior realização de cálculos.
df['cotacaoCompra'] = df['cotacaoCompra'].str.replace(',', '.').astype(float)
df['cotacaoVenda'] = df['cotacaoVenda'].str.replace(',', '.').astype(float)
df['dataHoraCotacao'] = pd.to_datetime(df['dataHoraCotacao'])
```

```python
#verificando a consitência e variabilidade dos dados
# Verificar valores nulos
print(df.isnull().sum())

# Resumo estatístico para verificar outliers
print(df.describe())
```

```python
# Por se trtar de cotação, interessante analisar as séries temporais - Gráfica de linhas é uma opção viável
#Tendo em vista que trabalhamos com as variáveis de venda e compra, interessante plotar 2 figuras distintas


# Gráfico para Cotação de Compra
plt.figure(figsize=(12, 6))
plt.plot(df['dataHoraCotacao'], df['cotacaoCompra'], label='Cotação Compra', color='blue')
plt.xlabel('Data')
plt.ylabel('Cotação (R$)')
plt.title('Flutuação da Cotação de Compra do Dólar')
plt.legend()
plt.grid()
plt.show()

# Gráfico para Cotação de Venda
plt.figure(figsize=(12, 6))
plt.plot(df['dataHoraCotacao'], df['cotacaoVenda'], label='Cotação Venda', color='orange')
plt.xlabel('Data')
plt.ylabel('Cotação (R$)')
plt.title('Flutuação da Cotação de Venda do Dólar')
plt.legend()
plt.grid()
plt.show()


```

### Pode parecer óbvio que os preços de venda e compra sejam muito próximos, contudo, isso indica uma economia bem regulada e com fatores de mercado organizado. O mesmo não ocorre em países com mercados de câmbio artificiais ou mecardo negro de moedas.

```python
# vamos analisar o spread, ou seja, a diferença entre compra e venda, e utilizar médias móveis, ou seja, cotações nas parciais semanais

# Garantir que as colunas são do tipo numérico e tratam string vazia
df['cotacaoCompra'] = pd.to_numeric(df['cotacaoCompra'], errors='coerce')
df['cotacaoVenda'] = pd.to_numeric(df['cotacaoVenda'], errors='coerce')

# Converter 'dataHoraCotacao' para datetime
df['dataHoraCotacao'] = pd.to_datetime(df['dataHoraCotacao'])

# Criar coluna 'data' (apenas a data, sem a hora)
df['data'] = df['dataHoraCotacao'].dt.date

# Spread entre cotação de compra e venda
df['spread'] = df['cotacaoVenda'] - df['cotacaoCompra']

# Média móvel de 7 dias
df['media_movel_compra'] = df['cotacaoCompra'].rolling(window=7).mean()
df['media_movel_venda'] = df['cotacaoVenda'].rolling(window=7).mean()
df['media_movel_spread'] = df['spread'].rolling(window=7).mean()

```

```python
# 1. Gráfico do Spread
plt.figure(figsize=(12, 6))
plt.plot(df['data'], df['spread'], label='Spread', color='green')  # Usando a coluna 'data'
plt.xlabel('Data')
plt.ylabel('Spread (R$)')
plt.title('Spread entre Cotação de Compra e Venda do Dólar')
plt.legend()
plt.grid()
plt.show()



# 2. Gráfico do Spread com Média Móvel do Spread
plt.figure(figsize=(12, 6))
plt.plot(df['data'], df['spread'], label='Spread')
plt.plot(df['data'], df['media_movel_spread'], label='Média Móvel do Spread (7 dias)', color='red')
plt.xlabel('Data')
plt.ylabel('Spread (R$)')
plt.title('Spread e Média Móvel do Spread da Cotação do Dólar')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calcular a diferença entre as médias móveis
df['diff_media_movel'] = df['media_movel_venda'] - df['media_movel_compra']

# Gráfico da Diferença entre as Médias Móveis
plt.figure(figsize=(12, 6))
plt.plot(df['data'], df['diff_media_movel'], label='Diferença (Venda - Compra)', color='purple')
plt.xlabel('Data')
plt.ylabel('Diferença (R$)')
plt.title('Diferença entre as Médias Móveis (7 dias) de Venda e Compra do Dólar')
plt.legend()
plt.grid()
plt.show()
```

## A Dança do Spread
### Os gráficos apontam para períodos de baixa diferenciação, contudo, tamvez seja o aspecto mais notável: o quão surpreendentemente houve estabilidade do  ao longo do período.

### O mercado demonstrou uma consistência notável em sua precificação. Mesmo nos momentos de maior volatilidade, como em 2017 e 2021, quando observamos picos isolados chegando a 0,51 centavos, o mercado rapidamente retornou aos seus patamares habituais.

```python
# Como forma de organizar os dados por período de Governo, vamos fazer uma análise por períodos de anos.# Criar um DataFrame com os mandatos dos presidentes

# Criar um DataFrame com os mandatos dos presidentes
data_presidentes = {
    'Presidente': ['Dilma Rousseff', 'Michel Temer', 'Jair Bolsonaro', 'Luiz Inácio Lula da Silva'],
    'Início': ['2015-01-01', '2016-08-31', '2019-01-01', '2023-01-01'],
    'Fim': ['2016-08-30', '2018-12-31', '2022-12-31', '2025-01-17'] #sua base contempla dados até essa data
}
df_presidentes = pd.DataFrame(data_presidentes)

# Converter as colunas 'Início' e 'Fim' para datetime
df_presidentes['Início'] = pd.to_datetime(df_presidentes['Início'])
df_presidentes['Fim'] = pd.to_datetime(df_presidentes['Fim'])

# Definindo 'Início' como índice para facilitar a busca
df_presidentes.set_index('Início', inplace=True)

print(df_presidentes)
```

```python
def get_presidente(data, df_presidentes):
    presidente = df_presidentes.loc[df_presidentes.index <= data]
    if not presidente.empty:
        return presidente.iloc[-1]['Presidente']
    else:
        return 'N/A'
```

```python
# Certifique-se de que a coluna 'data' esteja no formato datetime
df['data'] = pd.to_datetime(df['data'])

# Criando a coluna 'Presidente' usando a função apply e o DataFrame auxiliar
df['Presidente'] = df['data'].apply(lambda x: get_presidente(x, df_presidentes))

print(df)
```

```python
## Vamos analisar como ficou o dataset com as novas colunas
df.head(10)
```

```python
## Analisando o spread durante cada governo
presidentes = df['Presidente'].unique()

for presidente in presidentes:
    df_temp = df[df['Presidente'] == presidente]
    plt.figure(figsize=(12, 6))
    plt.plot(df_temp['data'], df_temp['spread'], label=f'Spread - {presidente}', color='green')
    plt.xlabel('Data')
    plt.ylabel('Spread (R$)')
    plt.title(f'Spread da Cotação do Dólar durante o Governo {presidente}')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Analisando agora as méidas móveis, utilizando o padrão semanal, ou seja, como se comportam as médias ao londo de cada grupo de 7 dias.
# Importante notar que cada governo é marcado não só por questões internas, como também externas
for presidente in presidentes:
    df_temp = df[df['Presidente'] == presidente]
    plt.figure(figsize=(12, 6))
    plt.plot(df_temp['data'], df_temp['media_movel_compra'], label=f'Média Móvel Compra - {presidente}', color='blue')
    plt.xlabel('Data')
    plt.ylabel('Cotação (R$)')
    plt.title(f'Média Móvel de Compra (7 dias) durante o Governo {presidente}')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
for presidente in presidentes:
    df_temp = df[df['Presidente'] == presidente]
    plt.figure(figsize=(12, 6))
    plt.plot(df_temp['data'], df_temp['media_movel_venda'], label=f'Média Móvel Venda - {presidente}', color='orange')
    plt.xlabel('Data')
    plt.ylabel('Cotação (R$)')
    plt.title(f'Média Móvel de Venda (7 dias) durante o Governo {presidente}')
    plt.legend()
    plt.grid()
    plt.show()
```

##Padrões de Estabilidade
#-------------------------------------------------------------
### A análise ano a ano revela um mercado cambial maduro e eficiente.
### As médias anuais do spread mantiveram-se notavelmente consistentes:
#------------------------------------------------------------
### 2015-2016: Média em torno de 0,61-0,62 centavos
### 2017-2018: Pequena variação para 0,62-0,61 centavos
### 2019-2020: Estabilidade em cerca de 0,61-0,60 centavos
### 2021-2022: Manutenção em aproximadamente 0,62-0,59 centavos
### 2023-2025: Consolidação em torno de 0,60 centavos

```python
# Agrupando por presidente e calculando a média da cotação do dia
# Usando um gráfico de barras fica mais fácil verifiacr visualmente o comportamento do dolar em cada governo.
df['cotacaoDia'] = (df['cotacaoCompra'] + df['cotacaoVenda']) / 2

cotacao_media_por_governo = df.groupby('Presidente')['cotacaoDia'].mean()

# Plotando o gráfico de barras
plt.figure(figsize=(12, 6))
cotacao_media_por_governo.plot(kind='bar', color='skyblue')
plt.xlabel('Presidente')
plt.ylabel('Cotação Média (R$)')
plt.title('Cotação Média do Dólar por Governo')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
